```python
import librosa
import librosa.display
import matplotlib.pyplot as plt
```

```python
# Load an audio file
file_path = r"C:\Users\44778\OneDrive\Desktop\UWE_Docs\7. Dissertation\Xeno Canto Audio Files\XC88143 - Little Spotted Kiwi - Apteryx owenii owenii.mp3"

# Load the audio file
audio_data, sample_rate = librosa.load(file_path, sr=None)  # sr=None to preserve original sample rate

# Plot the waveform
plt.figure(figsize=(10, 4))
librosa.display.waveshow(audio_data, sr=sample_rate)
plt.title('Waveform of Kiwi Call')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.show()
```

```python
# Extract MFCC features
mfccs = librosa.feature.mfcc(y=audio_data, sr=sample_rate, n_mfcc=13)

# Plot the MFCCs
plt.figure(figsize=(10, 4))
librosa.display.specshow(mfccs, sr=sample_rate, x_axis='time')
plt.colorbar()
plt.title('MFCCs of Kiwi Call')
plt.show()
```
