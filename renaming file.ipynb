```python
import os

def rename_audio_files(base_folder):
    # Traverse the base directory
    for subdir, dirs, files in os.walk(base_folder):
        folder_name = os.path.basename(subdir)
        
        # Filter only audio files (you can modify the extensions if necessary)
        audio_files = [f for f in files if f.endswith(('.mp3', '.wav', '.flac', '.aac'))]
        
        for idx, filename in enumerate(audio_files, start=1):
            file_extension = os.path.splitext(filename)[1]
            new_filename = f"{folder_name}_{idx}{file_extension}"
            old_file = os.path.join(subdir, filename)
            new_file = os.path.join(subdir, new_filename)
            
            os.rename(old_file, new_file)
            print(f"Renamed: {old_file} to {new_file}")

# Specify the base folder path
base_folder = r"C:\Users\44778\OneDrive\Desktop\UWE_Docs\7. Dissertation\Xeno Canto Audio Files\Other Birds"

# Call the function to rename files
rename_audio_files(base_folder)

```
